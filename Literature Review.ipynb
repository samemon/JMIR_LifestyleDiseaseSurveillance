## 1) An analysis of morbidity and mortality using Google Trends - Department of Economics, New Mexico State University (2017)

#### Main Idea:
    - Internet searches have become popular amongst patients.
    
    - This study: Investigate the possibility of using GT to see relation b/w internet searches in major causes of death and actual mortality rates in the U.S. 
    
    - Specifically looking at how GT performs in the context of top 10 causes of mortality in the US by CDC
    
    - Hypothesis: Google trends data on morbidity can be used to explain weekly mortality rates among specific age groups: 
        - Mortality_i = α + β1i SI coronary artery disease + β2i SI myocardial infarction + β3i SI stroke + β4i SI traffic collision + β5i SI nephritis + β6i SI suicide + β7i SI influenza + β8i SI viral pneumonia + β9i SI breast cancer + β10i SI Alzheimer’s + β11i SI diabetes + β12i SI chronic obstructive pulmonary disease + β13i SI prostate cancer + β14i SI leukemia + β15i weeks + εi
    
    
#### Method:
    - Least squares regression is used to estimate the relationships while accounting for a time trend. 
    
    - Weekly mortality is expressed as a function of search interests in the top ten causes of mortality in 2011 according to the CDC.
    
    - They handpicked some keywords for each disease
    
#### Results:
    - Significant relationships are found between search interests in seven of the top ten causes of mortality


#### Things that could be used (cited, compared) in our paper:
    - Mental Health Literature: Previous research have used Google Trends as a data source especially for mental health research purposes (Carneiro & Mylonakis, 2009; Eysenbach, 2006; Hulth & Rydevik, 2011; Kang, Zhong, He, Rutherford, & Yang, 2013; Sarigul & Rui, 2014; Pelat, Turbelin, Bar-Hen, Flahault, & Valleron, 2009)
    
    - Google queries related to other diseases (gastroenteritis, and chickenpox) have also been found to be correlated with clinical surveillance data from the French Sentinel Network (Pelat et al., 2009).
    
    - Google search volume index data has also been found useful in now- casting obesity across different metro areas (Sarigul & Rui, 2014) and has been used in the discussion of emerging diseases like Ebola (Milinovich, Magalhães, & Hu, 2015).
    
    - Google Trends data have also been used vastly in the study of mental illness, suicide and depression (McCarthy, 2010).
    
    - Stress related Google searches have been found to be positively and significantly related to suicide related searches and the actual incidence of suicide (Song, Song, An, Hayman, & Woo, 2014). 
    
    - Tobacco, Smoking and Alcohol related literature: Google Trends data can also be used to measure indicators of health behavior or the health interests of people including habits such as smoking and drinking. The use of tobacco among youth and adults has been investigated using Google Trends data and it was found that there exist a significant positive correlation between state level Google Trends cigar relative search volume and the prevalence of cigar use among youth and adults (Cavazos-Rehg et al., 2015). Drinking behavior measured by Google search data has also been found to be positively correlated to macroeconomic recessionary conditions (Frijters, Johnston, Lordan, & Shields, 2013).
    


## 2) Forecasting state-level premature deaths from alcohol, drugs, and suicides using Google Trends data - Michigan State University (2016)

#### Main Idea:
    - Rates of AICD, DICD, and suicides are only available within a 2 year lag.
    - Idea is to see if Google Trends can predict the variation in state-level rates in the US
    - Forecast the rates for 2015 (the data for which isn't available)
   
#### Method:
    - L1 regularization method using 27 key terms in contrast to the previous methods based on unemployment and income variables.
    - Used Google Trends Time Series Data for each term for each state
    - From the initial pool, we discarded 202 multi-word terms, and 121 terms with “too little search volume.”
    - Each term was carefully examined to consider possible confounding searches—searches for non-behavioral health terms using the same word—which could muddle the results. For instance, searching for “pot,” a common slang term for marijuana, also produces search volumes for phrases such as “crock pot,” a kitchen appliance.
    
#### Results:
    -  Google Trends data account for substantial variation in growth of statelevel rates of death: 30.9% for AICD, 23.9% for DICD, and 21.8% for suicide rates. Every state except Hawaii is forecasted to increase in all three of these rates in 2015.
    
#### Things that could be used (cited, compared) in our paper:
    - People are increasingly turning to the internet for information about health, mental health, and suicide - Fox, S., 2011. The Social Life o Health Information 2011. Pew Internet & American Life Project, Washington, DC.
    
    -  In Japan, Google searches for the term “depression” were positively correlated with the aggregate suicide rate after a 1–3 month lag (Sueki, 2011), while in Taiwan, nearly onethird of the variance in aggregated suicide rates were accounted for by searches for “major depression” and “divorce” (Yang et al., 2011).
    
    - In the United States, the terms “commit suicide,” “how to suicide,” and “suicide prevention” were all positively correlated with the nationwide suicide rate (Gunn III and Lester, 2013).
    
    - Monitoring social media and searches online have proven fruitful in tracking nonsuicidal self-injury (Bragazzi, 2014), recreational drug use (Deluca et al., 2012), non-cigarette tobacco use (Cavazos-Rehg et al., 2014), cannabis use (Steppan et al., 2013), non-medical use of Adderall (Hanson et al., 2013), and alcohol use (Frijters et al., 2013).
    
    - One previous study considered state-level suicide surveillance using Google Trends; the study was cross-sectional in nature, using one year of statelevel data (n=51) and 5 search terms (Ma-Kellams et al., 2016).
    
    -  Google Flu Trends was eventually shut down because the model persistently over-predicted the number of flu deaths (Lazer et al., 2014). Lazer et al. (2014) collected a four-part summary of the problems with Google Flu Trends which we work to avoid in our model. First, Google Flu Trends had a problem with transparency and replicability because the code for the system was proprietary to Google even if the Google Trends data was public. The model specified in this paper is provided in as much detail as possible, although the generating process for the Google Trends data is still somewhat propriety to Google.
    
    - This study limited Google Trends to the United States, and focuses on the state-level data for comparability to official mortality data. Google's search volume tool starts with data from 2004. We limit the data to the last five years (2010–2014) because internet use has steadily increased over the last 15 years (Perrin and Duggan, 2015)

#### Something I did not understand:
    To model patterns in suicide and other death rates, the data were pre-treated by applying the natural logarithm before first-differencing. This procedure induces stationarity in the data, a necessary condition for linear modeling (Núñez‐Antón and Zimmerman, 2000). This pretreatment transforms the death rates into growth rates of the death rates.
    

## 3) Is Google Trends a reliable tool for digital epidemiology? Insights from different clinical settings - Emergency Department, Academic Hospital of Parma, Parma, Italy (2017)

#### Main Idea:
    - Compare the reliability of Google Trends in Italy in different clinical settings, for both common diseases with lower media coverage, and for less common diseases attracting major media coverage.
    
#### Method:
    - Using the keywords ‘‘renal colic”, ‘‘epistaxis”, and ‘‘mushroom poisoning”, selected on the basis of available and reliable epidemiological data to look for rarely spoken of diseases.
    
    - Using ‘‘meningitis”, ‘‘Legionella Pneumophila pneumonia”, and ‘‘Ebola fever” for popularly spoken of diseases.

#### Results:

    - No correlation was found between data captured from Google Trends and epidemiology of renal colics, epistaxis and mushroom poisoning. Only when searching for the term ‘‘mushroom” alone the Google Trends search generated a seasonal pattern which almost overlaps with the epidemiological profile, but this was probably mostly due to searches for harvesting and cooking rather than to for poisoning.
    
    - The Google Trends data also failed to reflect the geographical and temporary patterns of disease for meningitis, Legionella Pneumophila pneumonia and Ebola fever.

#### Things that could be used (cited, compared) in our paper: 
    - This paper says that Google Trends has modest reliability for defining the epidemiology of relatively common diseases with minor media coverage, or relatively rare diseases with higher audience. Overall, Google Trends seems to be more influenced by the media clamor than by true epidemiological burden. We could use this paper as a citation as to why we removed the branded key terms.
    
    - The US Institute of Medicine (IOM) has also recently acknowledged that the use of internet data in health care research holds promise, and may also ‘‘complement and extend the data foundations that presently exist. (Barrett-Connor E, Ayanian JZ, Brown ER, et al. A Nationwide framework for surveillance of cardiovascular and chronic lung diseases. Washington (DC): National Academies Press (US); 2011.)
    
    - We can also use it to discuss on why google trends can't possibly meet the trivial baseline as it seems to be more influenced by media clamor than by the true epidemiological impact of disease, at least in the diseases examined.

## 1) An analysis of morbidity and mortality using Google Trends - Department of Economics, New Mexico State University (2017)

#### Main Idea:
    - Internet searches have become popular amongst patients.
    
    - This study: Investigate the possibility of using GT to see relation b/w internet searches in major causes of death and actual mortality rates in the U.S. 
    
    - Specifically looking at how GT performs in the context of top 10 causes of mortality in the US by CDC
    
    - Hypothesis: Google trends data on morbidity can be used to explain weekly mortality rates among specific age groups: 
        - Mortality_i = α + β1i SI coronary artery disease + β2i SI myocardial infarction + β3i SI stroke + β4i SI traffic collision + β5i SI nephritis + β6i SI suicide + β7i SI influenza + β8i SI viral pneumonia + β9i SI breast cancer + β10i SI Alzheimer’s + β11i SI diabetes + β12i SI chronic obstructive pulmonary disease + β13i SI prostate cancer + β14i SI leukemia + β15i weeks + εi
    
    
#### Method:
    - Least squares regression is used to estimate the relationships while accounting for a time trend. 
    
    - Weekly mortality is expressed as a function of search interests in the top ten causes of mortality in 2011 according to the CDC.
    
    - They handpicked some keywords for each disease
    
#### Results:
    - Significant relationships are found between search interests in seven of the top ten causes of mortality


#### Things that could be used (cited, compared) in our paper:
    - Mental Health Literature: Previous research have used Google Trends as a data source especially for mental health research purposes (Carneiro & Mylonakis, 2009; Eysenbach, 2006; Hulth & Rydevik, 2011; Kang, Zhong, He, Rutherford, & Yang, 2013; Sarigul & Rui, 2014; Pelat, Turbelin, Bar-Hen, Flahault, & Valleron, 2009)
    
    - Google queries related to other diseases (gastroenteritis, and chickenpox) have also been found to be correlated with clinical surveillance data from the French Sentinel Network (Pelat et al., 2009).
    
    - Google search volume index data has also been found useful in now- casting obesity across different metro areas (Sarigul & Rui, 2014) and has been used in the discussion of emerging diseases like Ebola (Milinovich, Magalhães, & Hu, 2015).
    
    - Google Trends data have also been used vastly in the study of mental illness, suicide and depression (McCarthy, 2010).
    
    - Stress related Google searches have been found to be positively and significantly related to suicide related searches and the actual incidence of suicide (Song, Song, An, Hayman, & Woo, 2014). 
    
    - Tobacco, Smoking and Alcohol related literature: Google Trends data can also be used to measure indicators of health behavior or the health interests of people including habits such as smoking and drinking. The use of tobacco among youth and adults has been investigated using Google Trends data and it was found that there exist a significant positive correlation between state level Google Trends cigar relative search volume and the prevalence of cigar use among youth and adults (Cavazos-Rehg et al., 2015). Drinking behavior measured by Google search data has also been found to be positively correlated to macroeconomic recessionary conditions (Frijters, Johnston, Lordan, & Shields, 2013).
    


## 2) Forecasting state-level premature deaths from alcohol, drugs, and suicides using Google Trends data - Michigan State University (2016)

#### Main Idea:
    - Rates of AICD, DICD, and suicides are only available within a 2 year lag.
    - Idea is to see if Google Trends can predict the variation in state-level rates in the US
    - Forecast the rates for 2015 (the data for which isn't available)
   
#### Method:
    - L1 regularization method using 27 key terms in contrast to the previous methods based on unemployment and income variables.
    - Used Google Trends Time Series Data for each term for each state.
    - From the initial pool, we discarded 202 multi-word terms, and 121 terms with “too little search volume.”
    - Each term was carefully examined to consider possible confounding searches—searches for non-behavioral health terms using the same word—which could muddle the results. For instance, searching for “pot,” a common slang term for marijuana, also produces search volumes for phrases such as “crock pot,” a kitchen appliance.
    
#### Results:
    -  Google Trends data account for substantial variation in growth of statelevel rates of death: 30.9% for AICD, 23.9% for DICD, and 21.8% for suicide rates. Every state except Hawaii is forecasted to increase in all three of these rates in 2015.
    
#### Things that could be used (cited, compared) in our paper:

    -  In Japan, Google searches for the term “depression” were positively correlated with the aggregate suicide rate after a 1–3 month lag (Sueki, 2011), while in Taiwan, nearly onethird of the variance in aggregated suicide rates were accounted for by searches for “major depression” and “divorce” (Yang et al., 2011).
    
    - In the United States, the terms “commit suicide,” “how to suicide,” and “suicide prevention” were all positively correlated with the nationwide suicide rate (Gunn III and Lester, 2013).

## 3) Is Google Trends a reliable tool for digital epidemiology? Insights from different clinical settings - Emergency Department, Academic Hospital of Parma, Parma, Italy (2017)

#### Main Idea:
    - Compare the reliability of Google Trends in Italy in different clinical settings, for both common diseases with lower media coverage, and for less common diseases attracting major media coverage.
    
#### Method:

    - Data used for spatio-temporal in a sense that this was focused on a specific province.
    - Using the keywords ‘‘renal colic”, ‘‘epistaxis”, and ‘‘mushroom poisoning”, selected on the basis of available and reliable epidemiological data to look for rarely spoken of diseases.
    
    - Using ‘‘meningitis”, ‘‘Legionella Pneumophila pneumonia”, and ‘‘Ebola fever” for popularly spoken of diseases.

#### Results:

    - No correlation was found between data captured from Google Trends and epidemiology of renal colics, epistaxis and mushroom poisoning. Only when searching for the term ‘‘mushroom” alone the Google Trends search generated a seasonal pattern which almost overlaps with the epidemiological profile, but this was probably mostly due to searches for harvesting and cooking rather than to for poisoning.
    
    - The Google Trends data also failed to reflect the geographical and temporary patterns of disease for meningitis, Legionella Pneumophila pneumonia and Ebola fever.

#### Things that could be used (cited, compared) in our paper: 
    - This paper says that Google Trends has modest reliability for defining the epidemiology of relatively common diseases with minor media coverage, or relatively rare diseases with higher audience. Overall, Google Trends seems to be more influenced by the media clamor than by true epidemiological burden. We could use this paper as a citation as to why we removed the branded key terms.
    
    - The US Institute of Medicine (IOM) has also recently acknowledged that the use of internet data in health care research holds promise, and may also ‘‘complement and extend the data foundations that presently exist. (Barrett-Connor E, Ayanian JZ, Brown ER, et al. A Nationwide framework for surveillance of cardiovascular and chronic lung diseases. Washington (DC): National Academies Press (US); 2011.)
    
    - We can also use it to discuss on why google trends can't possibly meet the trivial baseline as it seems to be more influenced by media clamor than by the true epidemiological impact of disease, at least in the diseases examined.

## 8) Relationship Between State-Level Google Online Search Volume and Cancer Incidence in the United States: Retrospective Study - (2015)

#### Main Idea: 

Characterize the relationship between cancer incidence and online Google search volumes in the United States for 6 common cancers. A secondary objective was to evaluate the association of search activity with cancer-related public events and celebrity news coverage.

#### Methods:

    - Data:
        - CDC State level data.
        - For the primary analysis, each of the 6 cancer search terms was used individually to obtain the annual RSV for each state from 2004 through 2013. 
            - For example, if Kentucky were the state that searched for “breast cancer” the most in a given year, it would have a value of 100, while Kansas would have a value of 50 if it searched for “breast cancer” half as often as Kentucky in that year. 
        -  Second, all cancer search terms were compared relative to each other for the United States as a whole for the study period, 2004 through 2013, to contextualize national trends for cancer type over time.
        - The secondary objective was to explore the impact of known drivers of Internet activity in our GT data. In this analysis, the RSV trends for each cancer were temporally compared to events previously reported to increase search activity, including cancer awareness months, celebrity events, and heavily covered news stories [19-22]. In this analysis, RSV was determined by setting 100 at the time period with greatest activity for the United States as a whole. Cancer awareness months for each of the 6 cancers of interest are October (breast), September (prostate), November (lung), March (colon), and September (uterine and leukemia). For other noticeable spikes in RSV, Google searches were performed in a 2-step process first using the cancer and date range to identify news stories for that timeframe: “lung cancer March 2010.” If that search was not productive, a second search with the term “celebrity” was added: “lung cancer March 2010 celebrity.”
        
    - Analysis:
        - Primary: Spearman correlation and linear regression with RSV and year as independent variables and cancer incidence as the dependent variable.
        - Secondary:  Temporal associations between search activity and events raising public awareness such as cancer awareness months and cancer-related celebrity news were described.
        
    - More details about analysis:
        - Primary: 
            - correlation between rsv and cancer incidence at state level from 2004 to 2013.
            - For each cancer they obtained 510 rsvs (10 per state for each year). Then they used spearman correlation to examine association b/w state level rsv and state level incidence.

## 10) Using Search Engines to Investigate Shared Migraine Experiences

#### Main Idea: 
    - To investigate migraine patterns in the United States using Google search data. 
    - Evaluate time-series relationships between migraines and social factors.
    -  Hypothesis: RSV for migraine would be affected by societal aspects such as day of the week, holidays, and novel social events.
    
#### Methods:
    - Temporal data: 12 years of Google Trends data (January 1, 2004 to August 15, 2016) daily.
    
#### Interesting results:
    - The first pattern is that migraine searches in the United States have been increasing in relative popularity (ie, after adjusting for simultaneous increases in search volume) since about 2010. The reason for this increase is unclear, but compared to before 2010, the relative search volume has increased each year. A second pattern is that there is a rhythm to the searches throughout the year. Migraine searches are at their nadir in December and increase remarkably in January through March where they then decrease in relative volume into the summer months. Searches increase again in relative volume near the end of summer and peak again in the fall (Sept.–Nov.). This pattern is remarkably consistent from year-to-year and only varies by degree.

## 12) A Google Trends-based approach for monitoring NSSI - (2014)

#### Main Idea: 
    - Create a surviellance system for NSSI
    
    
#### Method:
    - Google Trends was searched for the years 2004 to 2012 (as data before 2004 were not available) for “autolesionismo” (Italian for NSSI). 
    - Autocorrelation functions of this series of data, both crude and adjusted partial, were computed using R environment.
    - A list of NSSI-related terms was searched in Italian and their flux volumes were correlated with the NSSI hit-search data. Pearson’s coefficient was calculated with SPSS. P-value was computed with SPSS and values ≤0.05 were considered statistically significant.

#### Results:
    - The Italian towns with the higher searching fluxes and volumes were Naples, Milan, Padua, Bari, Bologna, Rome, Palermo, Turin, and Florence; that is to say, they did not observe any geographical gradient between North and South.


#### Interesting idea(s):
    - The new media can have a dangerous epidemic and “viral” contagious effect on vulnerable subjects, and some scholars talk about a so-called “virtual cutting edge” (the ways adolescents use the Internet for conveying their feelings and NSSI-related experiences), a “Werther effect” (that is to say, the effect of improper and inappropriate media reporting of suicide, especially among celebrities and stars, on the lay population), or a “cyber-suicide phenomenon” (that is to say, a suicide attempt first announced on Internet websites, such as forums, blogs, or chats, which is later followed by the fatality). NSSI 2.0, being interactive, dynamic, and increasingly expanding, is a complex reality where websites could have a negative effect of reinforcement, or, on the contrary, could be used for healthier purposes, such as health education and self-managing and coping strategy. Therefore, monitoring the hit searches could act as an efficient real time surveillance system.

## 13) Low validity of Google Trends for behavioral forecasting of national suicide rates - (2017)

#### Main idea:
    -  This study evaluated the validity and utility of Google Trends search volumes for behavioral forecasting of suicide rates in the USA, Germany, Austria, and Switzerland

#### Goals:
    - Systematically collect relevant search terms in the field of suicide research, both with regards to pro-suicide terms and suicide prevention terms, and to evaluate respective Google Trends search volumes for availability and suitability for time-series analysis.
    - Conduct rigorous analyses on the temporal associations of search volumes and suicide rates, utilizing Box-Jenkins methods of time-series analysis, transfer function models and prewhitening
    - Compare results for different countries, namely the USA, Germany, Austria, and Switzerland, in order to gain an impression of the generality of findings and their reproducibility across different countries, and to obtain first results for German-speaking countries for which evidence is currently lacking. 
    

#### Flaws with previous studies:
    - First, studies selected search terms mostly in an ad hoc fashion and did not systematically evaluate for which terms Google Trends actually provides analyzable data.
    - Second, assessing the temporal associations between two time series may result in spurious findings when the analysis does not adequately account for structural similarities (e.g., autocorrelational pattern, temporal trends) in the explanatory time series (i.e., Google Trends search volumes) and the dependent time series (i.e., suicide rates) - This I do not understand.
    
#### Keyword Selection:
    - Search terms consisting of more than one word were searched both with and without quotation marks. Search terms consisting of more than one word were also entered in different spellings and in different word orders to cover a broad spectrum of possibilities. Additional terms suggested by Google’s autocomplete function in these searches were amended to the list of search terms. In a last step, we excluded search terms that pertained to movie titles, band or song titles, terms that are used in a lyrical or humorous context, scientific terms, and opaque or rare phrases.
